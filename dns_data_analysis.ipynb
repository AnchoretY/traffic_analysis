```python
import pandas as pd
```

```python
df_dns = pd.read_csv("./data/dns_output/dns.csv")
df_file = pd.read_csv("./data/dns_output/file_explore.csv")
df_heartbeat1 = pd.read_csv("./data/dns_output/心跳包1.csv")
df_heartbeat2 = pd.read_csv("./data/dns_output/心跳包2.csv")
df_shell = pd.read_csv("./data/dns_output/shell_whoami.csv")
df_hashdump = pd.read_csv("./data/dns_output/hashdump.csv")
df_screenshot = pd.read_csv("./data/dns_output/screenshot.csv")
df_sleep = pd.read_csv("./data/dns_output/sleep_10.csv")
```

```python
cols = [
    "dns.qry.name","frame.len","frame.time_delta_displayed","dns.qry.type","response.frame.len","dns.a","dns.txt.length"
]
df_dns = df_dns[cols]
df_file = df_file[cols]
df_heartbeat1 = df_heartbeat1[cols]
df_heartbeat2 = df_heartbeat2[cols]
df_shell = df_shell[cols]
df_hashdump = df_hashdump[cols]
df_screenshot = df_screenshot[cols]
df_sleep = df_sleep[cols]
```

```python

def split_domain(domain):
    tmp = domain.split(".")
    top_three_domain = ".".join(tmp[-4:])
    other_domain = ".".join(tmp[:-4])
    return top_three_domain,other_domain

df_dns["top_four_domain"] = df_dns["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_dns["other_domain"] = df_dns["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_file["top_four_domain"] = df_file["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_file["other_domain"] = df_file["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_heartbeat1["top_four_domain"] = df_heartbeat1["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_heartbeat1["other_domain"] = df_heartbeat1["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_heartbeat2["top_four_domain"] = df_heartbeat2["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_heartbeat2["other_domain"] = df_heartbeat2["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_shell["top_four_domain"] = df_shell["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_shell["other_domain"] = df_shell["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_hashdump["top_four_domain"] = df_hashdump["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_hashdump["other_domain"] = df_hashdump["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_screenshot["top_four_domain"] = df_screenshot["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_screenshot["other_domain"] = df_screenshot["dns.qry.name"].apply(lambda x:split_domain(x)[1])

df_sleep["top_four_domain"] = df_sleep["dns.qry.name"].apply(lambda x:split_domain(x)[0])
df_sleep["other_domain"] = df_sleep["dns.qry.name"].apply(lambda x:split_domain(x)[1])
```
