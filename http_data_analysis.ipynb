```python
import pandas as pd
import numpy as np

def get_time_interval(df):
    resquest_time_l = df.request_time.values.tolist()
    response_time_l = [0]+df.response_time.values.tolist()
    response_time_l = response_time_l[:-1]
    df["time_interval"] = np.array(resquest_time_l)-np.array(response_time_l)
    return df
```
